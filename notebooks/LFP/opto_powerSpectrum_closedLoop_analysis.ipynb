### Power Spectrum and statistics for individual animals, closed-loop

```python
animal_list = [
    "Winnie",
    "Frodo",
    "Totoro",
    "Banner",
    "Odins",
    "Bilbo",
    "Wallie",
    "Olive",
    "Yoshi",
    # "control",
    # "transfected",
]
# animal_list = ["Totoro"]
# animal_list = ["Yoshi","Winnie"]
# animal_list = ["Wallie"]
# animal_list = ["Banner"]
phase_list = np.array([90, 180, 270, 360])
track_type = "lineartrack"
track_type = "early_wtrack"
dlc_pos = "wtrack" in track_type

phase_list = np.array(
    [
        90,
    ]
)
# track_type = "early_wtrack"

# phase_list = np.array([90,])
bootstrap_scores = True
suppression_scores = {}

score_fig, score_ax = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(12, 6))
for n_animal, animal in enumerate(animal_list):
    score_list = []
    for phase in phase_list:
        dataset_key = {
            "optogenetic_protocol": "phase_targeting",
            "laser_power": 77,
            "animal": animal,
            "min_pulse_length": 25,
            "max_pulse_length": 42,
            "targeted_phase": phase,
            "track_type": track_type,
        }

        fig, scores = opto_spectrum_analysis(dataset_key, return_distributions=True, dlc_pos=dlc_pos)
        score_list.append(list(scores))
        plt.rcParams["svg.fonttype"] = "none"
        if fig is not None:
            fig.savefig(
                f"./Figures_20241201/lfp_spectrum_{track_type}_phase{phase}_{animal}.svg"
            )

    score_list = np.array(score_list)
    suppression_scores[animal] = score_list[:, 0]
    score_names = ["suppression", "entrainment"]
    width = 20
    color = animal_style.loc[animal, "color"]  # plt.cm.jet(n_animal/8)
    for i in range(2):
        if bootstrap_scores:
            data = [bootstrap(dat, int(1e4)) for dat in score_list[:, i]]
        else:
            data = score_list[:, i].copy()
        violin = score_ax[i].violinplot(
            data,
            showmeans=False,
            showextrema=False,
            showmedians=False,
            positions=phase_list + n_animal * width / 2,
            widths=width,
        )
        for pc in violin["bodies"]:
            pc.set_facecolor(color)
            pc.set_alpha(0.3)
        score_ax[i].scatter(
            phase_list + n_animal * width / 2,
            [np.mean(d) for d in data],
            label=dataset_key["animal"],
            color=color,
        )
        score_ax[i].set_ylabel(score_names[i])
        score_ax[i].set_xticks(phase_list)
    score_ax[0].legend()
```

```python
import statsmodels.api as sms
import pandas as pd

model_df = []
for i, animal in enumerate(suppression_scores.keys()):
    data = suppression_scores[animal][0]  # [:,1][0]
    for d in data:
        model_df.append(
            {
                "Animal": i,
                "Treatment": (
                    "transfected"
                    if animal_style.loc[animal, "background"] == "indianred"
                    else "control"
                ),
                "Score": d,
            }
        )


model_df = pd.DataFrame(model_df)

from statsmodels.formula.api import mixedlm

# Convert the 'Treatment' column to a categorical variable
model_df["Treatment"] = model_df["Treatment"].astype("category")
# model_df['Animal'] = model_df['Treatment'].astype('category')

# Fit the mixed-effects model
model = mixedlm("Score ~ Treatment", model_df, groups=model_df["Animal"])
result = model.fit()

# Print the summary of the model
print(result.summary())
result.pvalues
```
